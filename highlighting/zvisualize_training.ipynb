```python
import plotly.express as px
import json
```

```python
with open("models/gpt2-xl-M1/training_details.json", "r") as f:
    tdata = json.load(f)
```

```python
tdata
```

```python
import numpy as np
```

```python
x = np.arange(0, len(tdata["loss_series"]))
xp = np.arange(1, len(tdata["val_series"]) + 1) * 300
yp = tdata["val_series"]
period = 3000/tdata["test_log_interval"]
valinter = np.interp(x, xp, yp)
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x = x,
        y = tdata["loss_series"],
        mode = "lines",
        name = "training loss"
    )
)
fig.add_trace(
    go.Scatter(
        x = x,
        y = valinter,
        mode = "lines",
        name = "validation loss"
    )
)
fig.show()
```
