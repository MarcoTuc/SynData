```python
import plotly.express as px
import json
import numpy as np
```

```python
model_path = "models/gpt2/M1"
with open(f"{model_path}/training_details.json", "r") as f:
    tdata = json.load(f)
```

```python
test_interval = tdata["test_log_interval"]
val_interval = tdata["train_size_aka_val_interval"]
batch_size = 10
```

```python
valscaling = val_interval / test_interval / batch_size
x = np.arange(0, len(tdata["loss_series"]))
xp = np.arange(1, len(tdata["val_series"]) + 1) * valscaling 
yp = tdata["val_series"]
period = 3000/tdata["test_log_interval"]
valinter = np.interp(x, xp, yp)
```

```python
import plotly.graph_objects as go
```

```python
val_series = tdata["val_series"]
```

```python
val_series
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x = x,
        y = tdata["loss_series"],
        mode = "lines",
        name = "training loss"
    )
)
fig.add_trace(
    go.Scatter(
        x = x,
        y = valinter,
        mode = "lines",
        name = "validation loss"
    )
)
fig.show()
```
